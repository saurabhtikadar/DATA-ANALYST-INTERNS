```python
import pandas as pd
```

```python
data= pd.read_csv('D:\DATA ANALYST\Internship\superstore.csv')
data
```

### Dropping The Row ID

The "Row ID" column is not really that informative, I think it would be safe enough for us to simply just delete them. That way, it would give us much more clarity over our dataset.

```python
data.drop(columns=['Row ID'], inplace=True) # removing the Row ID column as it is not needed
data
```

Now you can see, from the above table, we don't have the 'ROW ID' in place no longer. but column headers have space so we need to remove spaces.

```python
data.columns = data.columns.str.replace(' ', '_') # renaming columns to remove spaces
data
```

Now, you can see column headers have no spaces.Now we will check the data type of each column

```python
data.info()
```

In the above information you can see Order_Date and Ship_Date columns are currently of type Object.

Postal_Code, though numeric, should remain as Object to avoid misinterpretation (e.g., as a numerical value instead of a string).

Changes to Implement:
- OrderDate : *Object*
- ShipDate : *Object*
- PostalCode : *Object*

to

- OrderDate : *DateTime*
- ShipDate : *DateTime*
- PostalCode : *Object*

```python
data['Order_Date'] = pd.to_datetime(data['Order_Date'])
data['Ship_Date'] = pd.to_datetime(data['Ship_Date'])
data['Postal_Code'] = data['Postal_Code'].astype(str)
data.info()
```

Now, you can see datatype is changed.

```python
data.describe(include='all')
```

Now,We can use the **isnull()** method in Pandas to check for missing values in the dataset

```python
data.isnull()
```

Now we will remove the missing value

```python
data.dropna(inplace=True) # removing rows with missing values
data.isnull()
```

Now, missing value removed.

### Now we will create some new columns so that we can do the analysis in a better way

```python
data['Revenue'] = data['Sales'] - data['Discount']
data['Cost'] = data['Sales'] - data['Profit']
data
```

### Adding New Date-Based Columns

```python
data['Order Month'] = data['Order_Date'].dt.month # extracting month from Order Date
data['Order Year'] = data['Order_Date'].dt.year # extracting year from Order Date
data['Order Day of Week'] = data['Order_Date'].dt.dayofweek # extracting day of week from Order Date
data
```

# Now,will save it.

```python
data.to_csv('D:\DATA ANALYST\Internship\superstore_cleaned.csv', index=False)
```
